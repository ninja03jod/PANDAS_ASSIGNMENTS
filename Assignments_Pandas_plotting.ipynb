```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = np.random.random((4,4))
s
```

```python
df = pd.DataFrame(s)
df
```

```python
df.columns=["val_1","val_2","val_3","val_4"]
```

```python
df
```

```python
df.rename(columns={"val_2":"data_2"})
```

```python
# lets plot a bar plot first
df.plot(kind="bar",color=["c","m","r","g"])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar Plot")
```

```python
# lets seperate each columns plot
# we can convert it to rows and columns
df.plot(kind="bar",subplots=True,layout=(2,2));
```

```python
# lets add labels of bar
bar = plt.bar([0,1,2,3],[10,20,30,40])
plt.bar_label(bar,label_type="center");
```

```python
# select 2 columns :
df[["val_1","val_2"]].plot(kind="bar")
```

```python
df[["val_4"]].plot(kind='bar')
```

```python
# lets perform below operations on barh: horizontal bar
#- Select 2 columns
#- select 1 column
#- separate all plots
#- change layout 2,2
```

```python
df.plot(kind="barh")
```

```python
# select 2 columns :
df[["val_3","val_1"]].plot(kind='barh',color=["c","m"])
plt.yticks([0,1,2,3],[2020,2021,2022,2023]);
```

```python
# seperate each column plot:
df.plot(kind="barh",subplots=True,figsize=(8,10));
```

```python
# change layout:
df.plot(kind="barh",subplots=True,layout=(1,4),figsize=(15,4)); # layout = (1,4) 1 row , 4 columns
```
